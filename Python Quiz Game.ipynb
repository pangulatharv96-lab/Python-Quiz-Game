<h1>Quiz Game 

Objective: 
Create a command-line Quiz Game that asks multiple-choice questions, takes the user's answers, 
evaluates the score, and gives a performance review at the end. 
Features: 
1. Show multiple questions with 4 options each. 
2. Accept user input (A/B/C/D) and check if it‚Äôs correct. 
3. Keep track of the total score using a for loop. 
4. Show final score and a custom message based on performance.

<h3>Quiz Game Project

```python
# Command-line Quiz Game

questions = [
    {
        "question": "What does SQL stand for?",
        "options": {
            "A": "Structured Query Language",
            "B": "Simple Query Language",
            "C": "Standard Question Language",
            "D": "Sequential Query Logic"
        },
        "answer": "A"
    },
    {
        "question": "Which Excel function is used to count numbers?",
        "options": {
            "A": "COUNT()",
            "B": "SUM()",
            "C": "AVERAGE()",
            "D": "IF()"
        },
        "answer": "A"
    },
    {
        "question": "Which tool is used for data visualization?",
        "options": {
            "A": "MySQL",
            "B": "Power BI",
            "C": "Python",
            "D": "Excel"
        },
        "answer": "B"
    },
    {
        "question": "Which SQL clause is used to filter data?",
        "options": {
            "A": "GROUP BY",
            "B": "ORDER BY",
            "C": "WHERE",
            "D": "SELECT"
        },
        "answer": "C"
    }
]

score = 0

print("üìä Welcome to the Data Analyst Quiz!\n")

for i, q in enumerate(questions, start=1):
    print(f"Q{i}. {q['question']}")
    for option, value in q["options"].items():
        print(f"   {option}. {value}")

    user_answer = input("Your answer (A/B/C/D): ").upper()

    if user_answer == q["answer"]:
        print("‚úÖ Correct!\n")
        score += 1
    else:
        print(f"‚ùå Wrong! Correct answer is {q['answer']}\n")

print("üéØ Quiz Completed!")
print(f"Your Score: {score}/{len(questions)}")

# Performance Review
if score == len(questions):
    print("üåü Excellent! You got all answers right.")
elif score >= len(questions) // 2:
    print("üëç Good job! Keep practicing.")
else:
    print("üìò Needs Improvement. Practice more and try again!")
```
